```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix

from NN import nn

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
digits = load_digits()

X = digits.data # / 16
y = digits.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
structure = [64, 64, 10]
Ann = nn(structure)
```

```python
Ann.fit(X_train, y_train, 200, 0.001)
```

```python
y_pred = Ann.predict(X_test)
accuracy = Ann.score(y_pred, y_test)
print(f"My Neural Network Accuracy: {accuracy * 100:.2f}%")
```

```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', version=1, as_frame=False)
X = mnist.data
y = mnist.target.astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
structure = [784, 128, 64, 10] 
mnist_model = nn(structure)
```

```python
mnist_model.fit(X_train, y_train, 10, 0.001)
```

```python
y_pred = mnist_model.predict(X_test)
accuracy = mnist_model.score(y_pred, y_test)
print(f"My Neural Network Accuracy on mnist: {accuracy * 100:.2f}%")
```
